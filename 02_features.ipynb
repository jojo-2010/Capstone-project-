تحميل المكتبات

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

قراءة الملف

```python
df = pd.read_csv("cleaned_data.csv")

df
```

تحويل النصوص لأرقام

```python
df = pd.get_dummies(df, columns=["Neighborhood", "House Style"], drop_first=True)

df
```

تغيير القيم

```python
qual_map = {
    "Po":1, "Fa":2, "TA":3, "Gd":4, "Ex":5
}

df["Exter Qual"] = df["Exter Qual"].map(qual_map)

df
```

قسمة البيانات

```python
df["price_per_sqft"] = df["SalePrice"] / df["Gr Liv Area"]

df
```

ضرب البيانات

```python
df["quality_area"] = df["Overall Qual"] * df["Gr Liv Area"]

df
```

موازنة البيانات

```python
df["SalePrice_log"] = np.log1p(df["SalePrice"])

df
```

توحيد القيم

```python
scaler = StandardScaler()

df[["Gr Liv Area", "Total Bsmt SF"]] = scaler.fit_transform(
    df[["Gr Liv Area", "Total Bsmt SF"]]
)

df
```

طرح البيانات و تقسيمها الى فئات

```python
df["age"] = df["Yr Sold"] - df["Year Built"]

df["age_group"] = pd.cut(
    df["age"],
    bins=[0, 10, 30, 100],
    labels=["New", "Recent", "Old"]
)

df
```

اقوى الاعمدة تأثير

```python
corr = df.corr(numeric_only=True)

high_corr = corr["SalePrice"].abs().sort_values(ascending=False)

df
```

حفظ البيانات في ملف csv

```python
df.to_csv("featured_data.csv", index=False)

df
```

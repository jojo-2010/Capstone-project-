الحد الأعلى

```python
def clean_data(df):
    df = df.drop_duplicates()
#تم استخدام Ai في هذا الكود
    upper = df["SalePrice"].quantile(0.99)
    df["SalePrice"] = df["SalePrice"].clip(upper=upper)

    return df
```

```python
df = clean_data(df)

assert df["SalePrice"].min() > 0
assert df.shape[0] > 0
assert df.isnull().sum().sum() < 5000
```

تحميل الكتبات

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

قراءة الملف

```python
df = pd.read_csv("featured_data.csv")

df
```

انشاظ histplot

```python
sns.histplot(df["SalePrice"], kde=True)
plt.title("Distribution of Sale Price")
plt.show()
```

انساء boxblot

```python
sns.boxplot(x=df["Overall Qual"], y=df["SalePrice"])
plt.title("Price vs Quality")
plt.show()
```

انشاء scatterplot

```python
sns.scatterplot(x=df["Gr Liv Area"], y=df["SalePrice"])
plt.title("Area vs Price")
plt.show()
```

انشاء heatmap

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

متوسط السعر لكل عمر

```python
group = df.groupby("age_group")["SalePrice"].mean()
print(group)
```

متوسط السعر

```python
mean = np.sum(df["SalePrice"]) / len(df)
```

انتشار الاسعار

```python
std = np.sqrt(np.sum((df["SalePrice"] - mean)**2) / len(df))

print(mean, std)
```

بعد السعر عن المتوسط

```python
df["z"] = (df["SalePrice"] - mean) / std
```

احتمال كون البيت غالي

```python
high_price = df[df["SalePrice"] > 200000]
prob = len(high_price) / len(df)

print("Probability:", prob)
```
